**Qu'est-ce qu'un bot ?**

Un bot, il peut faire tout ce qu'on faites vous-même sur Discord comme poster ou supprimer des messages, ajouter des emoji en réaction à un message ou encore créer et changer les rôles des membres d'un serveur. Un bot est un agent logiciel qui fonctionne sur Internet et effectue des tâches répétitives qu'on lui a programmé. 

**Qu'est ce que le logging. À quoi servent les logs?**

Les logs (ou journaux) du serveur permettent, aux utilisateurs ayant la permission d'y accéder, d'obtenir l'ensemble des opérations effectuées sur un serveur. Ils permettent de surveiller ce qu'il se passe au cas où des actions malveillants ou faites par erreur survient pour pouvoir les annulés. Les bots  peuvent ensuite même bannir des membres au cas où ces derniers essaient de créer du spam ou faire des actions similaires.
